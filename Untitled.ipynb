```python
import numpy as np
from scipy.io import loadmat
from scipy.signal import butter, filtfilt
import pandas as pd
import os
import copy
from sklearn.model_selection import GroupShuffleSplit
from sklearn.preprocessing import StandardScaler,MinMaxScaler,Normalizer
from sklearn.utils import shuffle
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import layers
from tensorflow.keras.constraints import max_norm
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
def bandpass_filter(lowcut,highcut,fs,order = 4):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='bandpass')
    return b, a
def apply_filter(data,lowcut,highcut,fs=250):
    b,a = bandpass_filter(lowcut,highcut,fs)
    filtered_data = filtfilt(b,a,data)
    return filtered_data
```

```python
def create_dataset(elec_idx,data_length,folder_path):
    num_of_elec = len(elec_idx)
    X = np.array([],dtype=np.float32).reshape(0,len(elec_idx),data_length)
    for file_name in os.scandir(folder_path):
        file_content = loadmat(file_name.path)['data'][elec_idx,125:data_length+125,:,:]
        file_content = file_content.reshape(num_of_elec,data_length,240).transpose(2,0,1) # [samples,channels,timepoints]
        X = np.concatenate([X,file_content],axis=0)
    y = np.repeat(np.array([np.repeat(i,6) for i in range(40)],dtype=np.int32).reshape(1,240),35,axis=0).reshape(8400)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            X[i,j] = apply_filter(X[i,j],4,40)
    group_size = int(X.shape[0]/35)
    return X,y
    
```

```python
def split_data(X,y,data_length,num_channels):
    num_participants = 35
    participant_idx = np.repeat(np.arange(0,num_participants),X.shape[0]/num_participants)
    
    gss = GroupShuffleSplit(n_splits=1, test_size=1, random_state=42)
    train_val_idx,test_idx = next(gss.split(X,y,participant_idx))
    test_data = X[test_idx].reshape(-1,8,500,1)
    test_labels = y[test_idx]
    train_val_data = X[train_val_idx]
    train_val_labels = y[train_val_idx]
    gss = GroupShuffleSplit(n_splits=1, test_size=1, random_state=42)
    train_index, val_index = next(gss.split(train_val_data, train_val_labels, participant_idx[train_val_idx]))
    train_data = train_val_data[train_index].reshape(-1,8,500,1)
    train_labels = train_val_labels[train_index]
    val_data = train_val_data[val_index].reshape(-1,8,500,1)
    val_labels = train_val_labels[val_index]
    
    return train_data,train_labels,val_data,val_labels,test_data,test_labels
```

```python
def standardize(x_train,x_val,x_test,scaler_type='standard'):
    group_size = X_test.shape[0]
    if scaler_type == 'standard':
        scaler = StandardScaler()
    elif scaler_type == 'minmax':
        scaler = MinMaxScaler()
    for i in range(x_train.shape[1]):
        x_train[:, i,:,0] = scaler.fit_transform(x_train[:, i, :,0])
        x_test[:, i, :,0] = scaler.transform(x_test[:, i, :,0])
        x_val[:, i, :,0] = scaler.transform(x_val[:, i, :,0])
    return x_train,x_val,x_test
```

```python
X,y = create_dataset([52,54,55,56,58,59,60,61],500,'data/')#
X_train,y_train,X_val,y_val,X_test,y_test = split_data(X,y,500,8)
X_train,X_val,X_test = standardize(X_train,X_val,X_test,'standard')
```

```python
def create_model(lr = 0.001,patience = 10, units = 40,chan = 8):
    model = Sequential()
    model.add(layers.Conv2D(96,(1,256),padding = 'same',use_bias = False,input_shape=(chan,500,1)))
    model.add(layers.BatchNormalization())
    model.add(layers.DepthwiseConv2D((chan,1),depth_multiplier = 1, depthwise_constraint = tf.keras.constraints.max_norm(1.)))
    model.add(layers.BatchNormalization())
    model.add(layers.Activation('elu'))
    model.add(layers.AveragePooling2D((1,4)))
    model.add(layers.Dropout(0.5))                              
    model.add(layers.SeparableConv2D(96,(1,16), use_bias = False, padding = 'same'))                                  
    model.add(layers.BatchNormalization())
    model.add(layers.Activation('elu'))
    model.add(layers.AveragePooling2D((1,8)))                                 
    model.add(layers.Dropout(0.5))
    model.add(layers.BatchNormalization())
    model.add(layers.Flatten())
    model.add(layers.Dense(units = units))
    model.add(layers.Activation('softmax'))
    
    optimizer = tf.keras.optimizers.Adam(learning_rate = lr) 
    model.compile(loss='sparse_categorical_crossentropy',optimizer=optimizer, metrics=['accuracy'])
    earlystop = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0.0001,
    patience=patience,
    verbose=0)
    checkpoint = tf.keras.callbacks.ModelCheckpoint(
    filepath='./model5',
    monitor='val_loss',
    verbose=0,
    save_best_only=True,
    save_weights_only=False)
    return model, earlystop, checkpoint
```

```python

model,checkpoint,earlystop = create_model()
model.fit(X_train,y_train,batch_size = 16,epochs = 100,validation_data = (X_val,y_val),shuffle=True)        # , 
scores = model.evaluate(X_test, y_test, verbose=0)
print(f'Score for fold {1}: {model.metrics_names[0]} of {scores[0]}; {model.metrics_names[1]} of {scores[1]*100}%')

```
